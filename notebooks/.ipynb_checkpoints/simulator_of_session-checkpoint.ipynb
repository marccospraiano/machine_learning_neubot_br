## Adaptive Bitrate Algorithms

### Abr Base

```python
import numpy as np
from scipy import stats

# client side
class AbrBase(object):
    def __init__(self,name,bufferSize):
        self.name = name 
        self.bufferSize = bufferSize
        self.video = None 
        self.representation = None
        self.session = None

    def setVideo(self,video):
        self.video = video
        self.representation = np.array(video['representation'])

    def __predictThroughput(self):
        pass

    def adaptiveStrategy(self):
        pass
```

### Abr Optimized

```python
class AbrOptimized(AbrBase):
    def __init__(self,bufferSize=15):
        super().__init__('OP',bufferSize)

    def __predictThroughput(self,realDownthrpt): # trust throughput
        return realDownthrpt
    
    def adaptiveStrategy(self,realDownthrpt):

        predDownthrpt = self.__predictThroughput(realDownthrpt)
        # Based on the predicted throughput, the next line decides the rate for the next segment.
        index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
        # Here is simulated the download of a new segment
        elapsed = (self.representation[index,1])/(realDownthrpt*1000)
        bitrate = self.representation[index,0]
        return predDownthrpt,elapsed,bitrate
```

### Abr LEDBAT

```python
class AbrLedBat(AbrBase):
    def __init__(self,bufferSize=15):
        super().__init__('LB',bufferSize)
        
    def __predictThroughput(self,elapsed,previousDownthrpt):
        predDownthrpt = previousDownthrpt 
        if  elapsed > self.video['maxSegmentDuration']:
            rel_err = 1 - elapsed/self.video['maxSegmentDuration']
            predDownthrpt = predDownthrpt + rel_err * predDownthrpt
            predDownthrpt = max(self.representation[:,0].min(),predDownthrpt)
        return predDownthrpt
    
    def adaptiveStrategy(self,realDownthrpt,previousDownthrpt,elapsed):
        if(previousDownthrpt==None):
            index = 0
            predDownthrpt = 0
        else:
            predDownthrpt = self.__predictThroughput(elapsed,previousDownthrpt)
            # Based on the predicted throughput, the next line decides the rate for the next segment.
            index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
        # Here is simulated the download of a new segment
        elapsed = (self.representation[index,1])/(realDownthrpt*1000)
        bitrate = self.representation[index,0]
        return predDownthrpt,elapsed,bitrate
```

### Abr Rate Based with Harmonic Mean

```python
class AbrRateBasedHM(AbrBase):
    def __init__(self,bufferSize=15):
        super().__init__('RB',bufferSize)

    def __predictThroughput(self,listPreviousDownthrpt, wind = 5):
        return stats.hmean(listPreviousDownthrpt[-wind:])
    
    def adaptiveStrategy(self,realDownthrpt,listPreviousDownthrpt):
        if(len(listPreviousDownthrpt)==0):
            index = 0
            predDownthrpt = 0
        else:
            predDownthrpt = self.__predictThroughput(listPreviousDownthrpt)
            # Based on the predicted throughput, the next line decides the rate for the next segment.
            index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
        # Here is simulated the download of a new segment
        elapsed = (self.representation[index,1])/(realDownthrpt*1000)
        bitrate = self.representation[index,0]
        return predDownthrpt,elapsed,bitrate
```

### Abr Based on LSTNetBid

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.models import model_from_json
from __future__ import division
import logging
import os

log = logging.getLogger(__name__)
log.setLevel(logging.ERROR)

class Predictor:

  def __init__(self, f_model, f_json, custom_objects):
    self.model = self.__loadModel(f_model, f_json, custom_objects)

  def __loadModel(self, f_model, f_json, custom_objects):

    '''f_model: path to model.h5 file'''
    '''f_json: path to model.json file'''

    if f_model and f_json is not None:
      try:
        log.info("Loading model and weights ...")
      except NameError:
        pass
      
      # load and create model from json file
      if os.path.exists(f_json):
        with open(f_json, 'r') as json_file:
          try:
            log.debug("Loading model from: %s", f_json)
          except NameError:
            pass
            
          model = model_from_json(json_file.read(), custom_objects=custom_objects)
        
        # load weights from h5 file into new model
        if os.path.exists(f_model):
          try:
            log.debug("Loading weights from: %s", f_model)
          except NameError:
            pass
          
          model.load_weights(f_model)
        else:
          try:
            log.critical("File %s does not exist", f_model)
          except NameError:
            print("File %s does not exist" % (f_model))
      else:
        try:
          log.critical("File %s does not exist", f_json)
        except NameError:
          print("File %s does not exist" % (f_json))
    
    return model

  def predict(self,x_values):
    y_values = self.model.predict(x_values)
    return y_values

class AbrRateBasedLSTNet(AbrBase):
  def __init__(self, bufferSize=15, f_model=None, f_json=None, custom_objects=None):
    super().__init__('RB-LSTNet', bufferSize)
    self.model = self.__loadModel(f_model, f_json, custom_objects) if f_json else None;
  
  def __loadModel(self, model, json, c_objects):
    '''model: path to model file'''
    return Predictor(model, json, c_objects)

  def __predictThroughput(self,faturesValues):
    return self.model.predict(faturesValues)[0]
      
  def adaptiveStrategy(self,realDownthrpt,faturesValues):
    predDownthrpt = self.__predictThroughput(faturesValues)
    # Based on the predicted throughput, the next line decides the rate for the next segment.
    index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
    # Here is simulated the download of a new segment
    elapsed = (self.representation[index,1])/(realDownthrpt*1000)
    bitrate = self.representation[index,0]
    return predDownthrpt,elapsed,bitrate
```

### Abr Adaptec

```python
class AbrAdaptec(AbrBase):
  def __init__(self,bufferSize=15,bufferLow=5,bufferMin=2,c=1.2):
    super().__init__('AD',bufferSize)
    self.bufferLow = bufferLow
    self.bufferMin = bufferMin
    self.c = c
    
  def __predictThroughput(self,listPreviousDownthrpt,listPredDownthrpt, sigma=0.8):
    if(len(listPreviousDownthrpt)==1):
      return listPreviousDownthrpt[0]
    else:
      return sigma * listPredDownthrpt[-1] + (1-sigma)*listPreviousDownthrpt[-1]
    
  def adaptiveStrategy(self,realDownthrpt,listPreviousDownthrpt,
                       listPredDownthrpt,listRate,bufferOccupancy):
    if(len(listPreviousDownthrpt)==0):
      index = 0
      predDownthrpt = 0
    else:    
      predDownthrpt = self.__predictThroughput(listPreviousDownthrpt,listPredDownthrpt, sigma=0.8)
      if(bufferOccupancy > self.bufferLow and 8*predDownthrpt * self.c > listRate[-1] and listRate[-1] < self.representation[:,0][-1]):
          index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
          #index = np.array([i+1 if listRate[-1]==value else 0 for value,i in zip(representation[:,0],range(len(representation)))]).max()
      elif(bufferOccupancy < self.bufferLow and bufferOccupancy > self.bufferMin):
        if(listPreviousDownthrpt[-1]*self.c < listRate[-1] and listRate[-1] > self.representation[:,0][0]):
          index = np.array([i-1 if listRate[-1]==value else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
        elif(listPreviousDownthrpt[-1]*self.c > listRate[-1] and listRate[-1] < self.representation[:,0][-1]):
          index = np.array([i if (value/8)<=predDownthrpt else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max()
          #index = np.array([i+1 if listRate[-1]==value else 0 for value,i in zip(representation[:,0],range(len(representation)))]).max()
        else:
          index = np.array([i if listRate[-1]==value else 0 for value,i in zip(self.representation[:,0],range(len(self.representation)))]).max() 
      else:
        index = 0
    # Here is simulated the download of a new segment
    elapsed = (self.representation[index,1])/(realDownthrpt*1000)
    bitrate = self.representation[index,0]
    return predDownthrpt,elapsed,bitrate
```

## QoE Metrics

```python
class Metric:
  def __init__(self,name):
    self.name = name
  
  def calc(self,listRate,listRebuffer):
    pass

class MetricQoElin(Metric):
  def __init__(self):
    super().__init__('QoElin')
  
  def calc(self,listRate,listRebuffer):
    bitrateUtility = np.asarray(listRate).mean()
    rebufferingPenalty = 4.3*np.asarray(listRebuffer).mean()
    smoothnessPenalty = np.abs(np.asarray(listRate[1:])-np.asarray(listRate[:-1])).mean()
    qoe=np.asarray(listRate).sum()-4.3*np.asarray(listRebuffer).sum()-np.abs(np.asarray(listRate[1:])-np.asarray(listRate[:-1])).sum()
    return qoe,bitrateUtility,rebufferingPenalty,smoothnessPenalty

class MetricQoElog(Metric):
  def __init__(self):
    super().__init__('QoElog')

  def calc(self,listRate,listRebuffer):
    bitrateUtility = np.log(np.asarray(listRate)/np.asarray(listRate).min()).mean()
    rebufferingPenalty = 2.66*np.asarray(listRebuffer).mean()
    smoothnessPenalty = np.abs(np.log(np.asarray(listRate[1:])/np.asarray(listRate[1:]).min()) \
                  - np.log(np.asarray(listRate[:-1])/np.asarray(listRate[1:]).min())).mean()
    qoe=np.log(np.asarray(listRate)/np.asarray(listRate).min()).sum()\
                  -2.66*np.asarray(listRebuffer).sum()-np.abs(np.log(np.asarray(listRate[1:])/np.asarray(listRate[1:]).min()) \
                  - np.log(np.asarray(listRate[:-1])/np.asarray(listRate[1:]).min())).sum()
    return qoe,bitrateUtility,rebufferingPenalty,smoothnessPenalty


class MetricQoEhd(Metric):
  def __init__(self):
    super().__init__('QoEhd')

  def calc(self,listRate,listRebuffer):
    bitrateUtility = (np.asarray(listRate)*100).mean()
    rebufferingPenalty = 8.*(np.asarray(listRebuffer)).mean()
    smoothnessPenalty = np.abs((np.asarray(listRate[1:])*100)-(np.asarray(listRate[:-1])*100)).mean()
    qoe=(np.asarray(listRate)*100).sum()-8.*(np.asarray(listRebuffer)).sum()-np.abs((np.asarray(listRate[1:])*100)-(np.asarray(listRate[:-1])*100)).sum()
    return qoe,bitrateUtility,rebufferingPenalty,smoothnessPenalty
```

```python
import json
import time
import math
import logging
import numpy as np

import pandas as pd 

from scipy import stats
import matplotlib.patches as mpatches


sns.set(style="whitegrid")

logger = logging.getLogger(__name__)
logger.setLevel(logging.ERROR)
"'[CRITICAL:50;ERROR:40;WARNING:30;INFO:20;DEBUG:10;NOTSET:0]'" 

class Player:

  def __init__(self,session,servers,abrAlg=None):
    """
    session: session dictionary file path.
    servers: video dictionary file path.
    """

    logger.info('Starting client...')
    self.sessions = self.__loadSession(session)
    self.servers = self.__loadServers(servers)
    self.abrAlg = abrAlg
    self.metricQoE = None


  def __loadServers(self,servers):
    '''servers: path to videos.json file'''

    logger.info('Loading servers...')
    return Server(servers)

  def __loadSession(self,session):
    '''session: path to sessions.json file'''

    logger.info('Loading sessions...')
    with open(session, 'r') as f:
      s = json.load(f)
      return s
  
  def __conection(self,idSession,idVideo):
    '''idSession: id of session
       idVideo: id of video'''

    logger.info('Starting conection...')
    return self.sessions[idSession],self.servers.videos[idVideo]
  
  
  def startRealSession(self,idSession,idVideo,metricQoE):
    '''idSession: id of session
       idVideo: id of video'''
    logger.info('Starting session...')
    
    self.metricQoE = metricQoE
    session,video = self.__conection(idSession,idVideo)
    
    bufferOccupancy = 0
    accumulatedElapsed = []
    listRates = []
    lbufferOccupancy=[]
    downthrpts = []
    listRebuffer = []
    previousElapsed = session['connect_time'] # The elapsed time of establishing the connection
    for elapsed,downthrpt,rate in zip(session['elapsed'],session['downthrpt'],session['rate']):
      accumulatedElapsed.append(elapsed)
      listRates.append(rate/1000)
      listRebuffer.append(0 if (bufferOccupancy-elapsed)>0 else elapsed-bufferOccupancy)
      bufferOccupancy=(bufferOccupancy-elapsed)+video['maxSegmentDuration'] if len(lbufferOccupancy)>0 else video['maxSegmentDuration']  
      print("Downthrpt: {:7.3f}Mbps\tPredDownthrpt:{:7.3f}Mbps\tBitrate:{:7.3f}Mbps\tDownload_time:{:7.3f}s\tBuffer:{:7.3f}s\tRebuffer:{:7.3f}s"
            .format(downthrpt*8/1000,downthrpts[-1]*8/1000 if len(downthrpts)>0 else 0,rate/1000,elapsed,bufferOccupancy,listRebuffer[-1]))
      lbufferOccupancy.append(bufferOccupancy)
      downthrpts.append(downthrpt)
      previousElapsed = accumulatedElapsed[-1]
    QoE= self.metricQoE.calc(listRates,listRebuffer) 
    print("QoE: ",QoE)
    return accumulatedElapsed,[value*8/1000 for value in downthrpts],listRates,lbufferOccupancy


  def startSimSession(self,idSession,idVideo,metricQoE):
    '''idSession: id of session
       idVideo: id of video'''

    self.metricQoE = metricQoE
    s,video = self.__conection(idSession,idVideo)
    representation = np.array(video['representation'])
    self.abrAlg.setVideo(video)
    bufferSize = self.abrAlg.bufferSize
    
    lbufferOccupancy = []
    listRate = []
    accumulatedElapsed = []
    listPredDownthrpt = []
    listRebuffer = []
    listRealDownthrpt = s['downthrpt'] 
    bufferOccupancy = 0
    elapsedWithOutDown = 0
    elapsed = 0
    i=0
    while i < int(math.ceil(video['mediaPresentationDuration']/video['maxSegmentDuration'])):
        if bufferOccupancy+video['maxSegmentDuration'] <= bufferSize: # if the buffer is not full
            if(self.abrAlg.name=='LB'):
                previousDownthrpt = listRealDownthrpt[i-1] if i > 0 else None
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i],previousDownthrpt,elapsed)
            elif(self.abrAlg.name=='OP'):
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i])
            elif(self.abrAlg.name=='RB'):
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i],listRealDownthrpt[:i])
            elif(self.abrAlg.name=='AD'):
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i],listRealDownthrpt[:i],
                                                                       listPredDownthrpt,listRate,bufferOccupancy)
            elif(self.abrAlg.name=='RB-LSTNet'):
                """
                features input for model:
                ['minute','day','hour','iteration',
                'request_ticks','delay','delta_sys_time',
                'connect_time','delta_user_time','rate',
                'received','tcp_mean_wind','elapsed','downthpt']
                """
                featuresValues = np.asarray([s['connect_time'],
                                             s['day'][i],
                                             s['delay'][i],
                                             s['delta_sys_time'][i],
                                             s['delta_user_time'][i],
                                             s['elapsed'][i],
                                             s['hour'][i],
                                             s['iteration'][i],
                                             s['minute'][i],
                                             s['rate'][i],
                                             s['received'][i],
                                             s['request_ticks'][i],
                                             s['tcp_mean_wind'][i]])
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i],np.reshape(featuresValues, (featuresValues.shape[0], 1, featuresValues.shape[1])))
            
            elif(self.abrAlg.name=='AD-LSTNet'): 
                featuresValues = np.asarray([s['connect_time'],
                                       s['day'][i],
                                       s['delay'][i],
                                       s['delta_sys_time'][i],
                                       s['delta_user_time'][i],
                                       s['elapsed'][i],
                                       s['hour'][i],
                                       s['iteration'][i],
                                       s['minute'][i],
                                       s['rate'][i],
                                       s['received'][i],
                                       s['request_ticks'][i],
                                       s['tcp_mean_wind'][i]])
                predDownthrpt,elapsed,bitrate = self.abrAlg.adaptiveStrategy(listRealDownthrpt[i],listRealDownthrpt[:i],
                                                                       np.reshape(featuresValues, (featuresValues.shape[0], 1, featuresValues.shape[1])),listRate,bufferOccupancy)
            listPredDownthrpt.append(predDownthrpt)
            listRebuffer.append(0 if (bufferOccupancy-elapsed)>0 else elapsed-bufferOccupancy)
            bufferOccupancy = (bufferOccupancy-elapsed)+video['maxSegmentDuration'] if i>0 else video['maxSegmentDuration']
            lbufferOccupancy.append(bufferOccupancy)
            accumulatedElapsed.append(accumulatedElapsed[-1]+elapsedWithOutDown+elapsed if len(accumulatedElapsed) > 0 else elapsed)
            listRate.append(bitrate)
            elapsedWithOutDown = 0
            print("Downthrpt: {:7.3f}Mbps\tPredDownthrpt:{:7.3f}Mbps\tBitrate:{:7.3f}Mbps\tDownload_time:{:7.3f}s\tBuffer:{:7.3f}s\tRebuffer:{:7.3f}s"
                .format(listRealDownthrpt[i]*8/1000,listPredDownthrpt[i]*8/1000,listRate[i]/1000,elapsed,lbufferOccupancy[i],listRebuffer[i]))
            i+=1
        else: # if the buffer is full
            bufferOccupancy = (bufferOccupancy-video['maxSegmentDuration'])
            elapsedWithOutDown += video['maxSegmentDuration']
    QoE,bitrateUtility,rebufferingPenalty,smoothnessPenalty= self.metricQoE.calc([value/1000 for value in listRate],listRebuffer) 
    metricsQoE={'QoE':QoE,'bitrateUtility':bitrateUtility,'rebufferingPenalty':rebufferingPenalty,'smoothnessPenalty':smoothnessPenalty}
    print("QoE: {}\nBitrate Utility: {}\nRebuffering Penalty: {}\nSmoothness Penalty: {}".format(QoE,bitrateUtility,rebufferingPenalty,smoothnessPenalty))
    return accumulatedElapsed,[value*8/1000 for value in listRealDownthrpt],[value*8/1000 for value in listPredDownthrpt],[value/1000 for value in listRate],lbufferOccupancy,listRebuffer,metricsQoE
  
  def simBatch(self,idVideo,metricQoE):
    listQoE = []
    listBitrateUtility = []
    listRebufferingPenalty = []
    listSmoothnessPenalty = []
    for s in range(len(self.sessions)):
      _,_,_,_,_,_,metricsQoE=self.startSimSession(s,idVideo,metricQoE)
      listQoE.append(metricsQoE['QoE'])
      listBitrateUtility.append(metricsQoE['bitrateUtility'])
      listRebufferingPenalty.append(metricsQoE['rebufferingPenalty'])
      listSmoothnessPenalty.append(metricsQoE['smoothnessPenalty'])
    return np.asarray(listQoE),np.asarray(listBitrateUtility),np.asarray(listRebufferingPenalty),np.asarray(listSmoothnessPenalty) 
```
